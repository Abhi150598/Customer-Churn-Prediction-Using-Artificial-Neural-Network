```python
# Get summary statistics of the dataset
print(data.describe())
```

```python
# Visualize feature distributions
fig, axes = plt.subplots(3, 2, figsize=(24, 18))
axes = axes.flatten()
features = ["Geography", "Gender", "Tenure", "NumOfProducts", "HasCrCard", "IsActiveMember"]
for i in range(len(features)):
    sns.countplot(x=features[i], data=data, palette="Set2", ax=axes[i])
plt.show()
```

```python
# Visualize the target variable distribution
sns.countplot(x="Exited", data=data, palette="Set2")
plt.show()
```

```python
print("Gender Count :")
print(data['Gender'].value_counts())
print("\n")
print("Female Quit Rate :")
print(data[data["Gender"]=="Female"]["Exited"].value_counts())
print("\n")
print("Male Quit Rate :")
print(data[data["Gender"]=="Male"]["Exited"].value_counts())
```

It indicates a slightly higher proportion of male customers in the dataset.
Out of 4543 female customers, 1139 have churned (Exited = 1), while 3404 have not churned (Exited = 0).
The female quit rate is approximately 25.1%.
Out of 5457 male customers, 898 have churned (Exited = 1), while 4559 have not churned (Exited = 0).
The male quit rate is approximately 16.5%.

```python
sns.countplot(data=data, x="Gender", hue="Exited", palette="pastel")
```

```python
print("Adresses :")
print(data['Geography'].value_counts())
print("\n")
l =["France","Spain","Germany"]
for x in l:
    print(f"{x} Quit Rate :")
    print(data[data["Geography"]==x]["Exited"].value_counts())
    print("\n")
```

The quit rate for customers from France is approximately 16.2%.
The quit rate for customers from Spain is approximately 16.7%.
The quit rate for customers from Germany is approximately 32.5%.

```python
sns.countplot(data=data, x="Geography", hue="Exited", palette="pastel")
```
